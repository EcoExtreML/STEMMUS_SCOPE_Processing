```python
import xarray as xr
ds = xr.open_dataset("C:/STEMMUS_SCOPE_data/global/co2/CAMS_CO2_2003-2020_NW.nc")
# ds = ds.drop_duplicates(["latitude", "longitude"])
ds = ds.sel(latitude=52, longitude=5, method='nearest')
```

```python
ds.co2.sel(time=slice("2016", "2017")).plot()
```

```python
ds_sel = ds.sel(time=slice("2016", "2017"))
ds_sel = ds_sel.resample(time="1800S").interpolate('linear')
ds_sel = ds_sel.drop(['latitude', 'longitude'])
```

```python
ds_sel.sel(time=slice("2016-01-01","2016-01-02")).co2.plot(marker='.')
```
